## 数据类型与缺失值

上面都是non-null，不一定是没有缺失值（np.nan），可能是缺失值的标记符号不一样，查看上面的数据，中间有很大部分是Not Available，所以Not Available应该就是缺失值

```python
# Replace all occurrences of Not Available with numpy not a number
data = data.replace({'Not Available':np.nan})
```
